```python
hero_stats_df.info()
```

```python
hero_stats_df.describe()
```

```python
# combat
plt.figure(figsize=(12,4))
g = sns.barplot(y='combat', x='name', data=hero_stats_df.sort_values(by=['combat']))
g.set_xticklabels(labels=hero_stats_df.sort_values(by=['combat'])['name'], rotation=30)
g.set_xlabel('Hero')
g.set_ylabel('Combat')
```

```python
# durability
plt.figure(figsize=(12,4))
g = sns.barplot(y='durability', x='name', data=hero_stats_df.sort_values(by=['durability']))
g.set_xticklabels(labels=hero_stats_df.sort_values(by=['durability'])['name'], rotation=30)
g.set_xlabel('Hero')
g.set_ylabel('Durability')
```

```python
# Intelligence
plt.figure(figsize=(12,4))
g = sns.barplot(y='intelligence', x='name', data=hero_stats_df.sort_values(by=['intelligence']))
g.set_xticklabels(labels=hero_stats_df.sort_values(by=['intelligence'])['name'], rotation=30)
g.set_xlabel('Hero')
g.set_ylabel('Intelligence')
```

```python
# Speed
plt.figure(figsize=(12,4))
g = sns.barplot(y='speed', x='name', data=hero_stats_df.sort_values(by=['speed']))
g.set_xticklabels(labels=hero_stats_df.sort_values(by=['speed'])['name'], rotation=30)
g.set_xlabel('Hero')
g.set_ylabel('Speed')
```

```python
# Strength
plt.figure(figsize=(12,4))
g = sns.barplot(y='strength', x='name', data=hero_stats_df.sort_values(by=['strength']))
g.set_xticklabels(labels=hero_stats_df.sort_values(by=['strength'])['name'], rotation=30)
g.set_xlabel('Hero')
g.set_ylabel('Strength')
```

```python

def get_total_rating(row):
    traits = ['combat', 'durability', 'intelligence', 'power', 'speed', 'strength']
    mean = sum([row[t] for t in traits])/len(traits)
    return mean

hero_stats_df['total_rating'] = hero_stats_df.apply(get_total_rating, axis=1)
```

```python
# Total Rating
plt.figure(figsize=(15,8))
g = sns.barplot(y='total_rating', x='name', data=hero_stats_df.sort_values(by=['total_rating']))
g.set_xticklabels(labels=hero_stats_df.sort_values(by=['total_rating'])['name'], rotation=30)
g.set_xlabel('Hero')
g.set_ylabel('Total Rating')
g.set_title('Total Rating of Heros')
```

```python
ax = sns.boxplot(data=hero_stats_df, orient="h", palette="Set2")
ax.set_title('Variability of Ratings')
```
